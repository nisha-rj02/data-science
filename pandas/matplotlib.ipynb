###pie

```python
x =np.array([1,2,3,4,])
plt.pie(x)
plt.show()
```

```python
x=np.array([1,2,3,4])
label=np.array(["o2","Co2","N","others"])
plt.pie(x, labels=label,startangle=90)
plt.show()
```

```python
x = np.array([1,2,3,4,5])
gases=np.array(["Nitrogen","Oxygen","Carbon","Argon","Other"])
myexplode =[0.2,0,0,0,0]
plt.pie(x, labels=gases,startangle=90,explode=myexplode,shadow=True)
plt.show()
```

```python
x = np.array([1,2,3,4,5])
gases=np.array(["Nitrogen","Oxygen","Carbon","Argon","Other"])
myexplode =[0.2,0,0,0,0]
color = ["red","green","pink","yellow","grey"]
plt.pie(x, labels=gases,startangle=90,explode=myexplode,shadow=True,colors=color)
plt.show()
```
