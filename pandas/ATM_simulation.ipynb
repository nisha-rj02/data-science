```python
# ATM Simulation 

User_Pin = "1234"
User_Balance = 10000.0



def check_balance(balance):
    print(f"\nYour current bank balance is {balance} rupees")

def deposit(balance):
    amount = float(input("\nEnter amount to be depositted: "))
    if amount <= 0:
        print("Invalid amount...")
    else:
        balance += amount
        print(f"{amount} rupees deposited successfully!")
    return balance    
     

def withdraw(balance):
    amount = float(input("\nEnter amount to be withdrawn:"))
    if amount <= 0:
        print("Invalid amount...")
    elif amount > balance:
        print("Insufficient balance...")
    else:
        balance -= amount
        print(f"{amount} rupees withdrawn successfully!")
    return balance

def ATM_menu():
        print("ATM Menu")
        print("1-Check Your Balance")
        print("2-Deposit Money") 
        print("3-Withdraw Money")
        print("4-Exit")

def ATM_simulation():

    global User_Balance
    print("Welcome")
    pin = input("Please enter your 4-digit PIN: ")

    if pin != User_Pin:
        print("Incorrect PIN.Please enter valid pin!")
        return
    else:
        print("Login Successful!")

    while True:
        
        ATM_menu()
        x = input("Choose an option (1-4): ")

        if x == '1':
            check_balance(User_Balance)
        elif x == '2':
            User_Balance = deposit(User_Balance)
        elif x == '3':
            User_Balance = withdraw(User_Balance)
        elif x == '4':
            print("Thank you for using the ATM. Have a nice day!")
            break
        else:
            print("Invalid option. Please select a valid option.")

ATM_simulation()

```
